## Summary

The United States Bureau of Labor Statistics (BLS) surveys and calculates the monthly unemployment rate.
This notebook explores and visualizes the uenmployment rate data.

```python
# import necessary packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
# test if matplotlib works on jupyter
plt.plot()
plt.show()
```

```python
unrate = pd.read_csv("unrate.csv", sep=",")
```

```python
# explore unrate dataset
print(unrate.info())
```

```python
print(unrate.columns)
```

```python
print(unrate.head())
```

```python
# convert DATE column in string to datetime-format
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
# check if DATE column is now a datetime object
print(unrate.info())
```

```python
# line plot unemployment rate data 1948 to 2916
print(unrate.columns)
X = unrate['DATE']
Y = unrate['VALUE']
plt.plot(X,Y)
plt.show()
```

```python
# rotate x_ticks
plt.plot(X,Y)
plt.xticks(rotation= 90)
plt.show()
```

```python
#  add titles

plt.plot(X,Y)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948-2016")
plt.xticks(rotation= 90)
plt.show()
```

```python
# line plot unemployment rate data only 1948
X_48 = unrate.loc[0:11, 'DATE']
Y_48 = unrate.loc[0:11, 'VALUE']
plt.plot(X,Y)
plt.show()
```

```python
# rotate x_ticks for 1948
plt.plot(X_48,Y_48)
plt.xticks(rotation= 90)
plt.show()
```

```python
#  add titles
plt.plot(X_48,Y_48)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.xticks(rotation= 90)
plt.show()
```
